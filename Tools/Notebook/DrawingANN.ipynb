# Drawing ANN

## Importing library

```python
import matplotlib.pyplot as plt
from nnv import NNV
```

## Change the figure size

```python
plt.rcParams["figure.figsize"] = (10,10)
```

## Two layer ANN

```python

TwoLayers = [
    {"title":"input\n(relu)", "units": 3,  "color": "orange", "edges_color":"red", "edges_width":2},
    {"title":"output\n(sigmoid)", "units": 1,"color": "blue"},
]

NNV(TwoLayers, node_radius=10, spacing_layer=60, font_size=22).render(save_to_file="TwoLayers.png")
```

## Make the figure size larger

```python
plt.rcParams["figure.figsize"] = (100,10)
```

## Three layer ANN

```python
ThreeLayers = [
    {"title":"input\n(relu)", "units": 3, "color": "yellow"},
    {"title":"hidden\n(relu)", "units": 4, "color": "orange", "edges_color":"red", "edges_width":2},
    {"title":"output\n(sigmoid)", "units": 1,"color": "blue"},
        
]


NNV(ThreeLayers, max_num_nodes_visible=8, node_radius=10, spacing_layer=60, font_size=24).render(save_to_file="my_.pdf")
```

#### Example #3

```python
nUnits = 4
nLayers = 4
ManyLayers = []

ManyLayers.append({"title":f"input layer\n{nUnits} units\n(relu)", "units": 3, "color": "yellow"})
for i in range(1,nLayers-1):
    ManyLayers.append({"title":f"hidden layer {i}\n{nUnits} units\n(relu)", "units": nUnits, "color": "orange"})
ManyLayers.append({"title":"output layer\n1 unit\n(sigmoid)", "units": 1,"color": "blue"})
    
NNV(ManyLayers, max_num_nodes_visible=6, spacing_layer=90, font_size=28).render()
```
